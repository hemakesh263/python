# Factorial of a number

```python
num=int(input("enter a number"))
factorial=1
if num>0:
  for i in range(1,num+1):
    factorial=factorial*i
    print(factorial)
elif num==0:
  print("factorial of 0 is 1")
else:
  print("the given number is negative")
```

# Palindrome number or not

```python
n=int(input("enter a number:"))
rev=0
org=n
while (n!=0):
  remainder=n%10
  rev=rev*10+remainder
  n=int(n/10)
if (org==rev):
  print("palindrome")
else:
  print("not palindrome")
```

## SORT IN LISTS

### ASCENDING ORDER

```python
fruit=["apple","banana","orange","guava","grapes"]
fruit.sort()
print(fruit)
```

```python
list=[65,95,15,26,84,1,85,2,100,955]
list.sort()
print(list)
```

### DESCENDING ORDER

```python
list=[65,95,15,26,84,1,85,2,100,955]
list.sort(reverse=True)
print(list)
```

```python
fruit=["apple","banana","orange","guava","grapes"]
fruit.sort(reverse=True)
print(fruit)
```

### CASE INSENSITIVE SORT

```python
fruit=["apple","banana","Orange","guava","grapes"]
fruit.sort()
print(fruit)
```

```python
fruit=["apple","banana","Orange","guava","grapes"]
fruit.sort(key= str.lower)
print(fruit)
```

```python
fruit=["apple","banana","Orange","guava","grapes"]
fruit.reverse()
print(fruit)
```
