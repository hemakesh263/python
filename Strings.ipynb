# Strings

```python
a="Hemakesh"
print(a)
print(type(a))
print(a[3])
```

## Length of string

```python
a="Hemakesh Surla"
print(len(a))
```

### Finding something in a string

```python
a="RRR movie has been released"
print("RRR" in a)
```

```python
a="RRR movie has been released"
print("RRR" not in a)
```

### Slicing

```python
a="RRR movie has been released"
print(a[2:5])
```

## Modify string

```python
a="RRR movie has been released"
print(a.upper())
```

```python
a="RRR movie has been released"
print(a.lower())
```

### Remove whitespace


```python
a="    RRR movie has been released         "
print(a.strip())
```

### Replace string

```python
a="RRR movie has been released"
print(a.replace("RRR","Sarkaru vaari paata"))
```

### Split string

```python
a="RRR movie has been released"
print(a.split("has"))
```

```python
a="RRR movie has been released"
print(a.split())
```

## String Concatenation

```python
a="RRR movie has been released"
b="and fans are excited about it"
c=a+b
print(c)
```

```python
a="RRR movie has been released"
b="and fans are excited about it"
c=a+" "+b
print(c)
```

```python
a="Radhe Shyam is releasing on"
b="1st of april in amazon prime"
c=a+" "+b
print(c)
```

```python
a="neeku enduku"
b=12345
c=a+b
print(c)
```

## String format

```python
date=1
a="radheshyam is releasing on april {}"
print(a.format(date))
```

```python
price=500
quantity=2
itemno=231
customer="i want to pay {} for these {} items of {}"
print(customer.format(price,quantity,itemno))
```

```python
date=26
year=2006
me="pokiri has released on {} april {}"
print(me.format(date,year))
```

```python
shirt=1
price=6000
me="i have spent {1} for {0} shirt"
print(me.format(shirt,price))
```
