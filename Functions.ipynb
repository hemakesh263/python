# ARGUMENTS

```python
def hemakesh(name):
  print( "Super Star " + name)

hemakesh("Mahesh Babu")
hemakesh("Rajinikanth")
```

```python
def hemakesh(number):
  print("Pradeep " + number)
hemakesh("1233")
```

# NUMBER OF ARGUMENTS

```python
def hemakesh(movie,actor):
  print(movie + " has been played by " + actor)

hemakesh("Pokiri","Mahesh Babu")
```

# ARBITRARY ARGUMENTS, *args

```python
def hemakesh(*name):
  print("My favorite singer is " + name[1])

hemakesh("Darshan Raval","Arjit Singh","Sid Sriram")
```

# KEYWORD ARGUMENTS

```python
def hemakesh(name1,name2,name3,name4):
  print("The king of KGF is " + name4)

hemakesh(name1 = "Surya Vardhan",name2 = "Garuda",name3 = "Adheera",name4 = "Rocky")
```

# ARBITRARY KEYWORD ARGUMENTS, *kwargs

```python
def hemakesh(**name):
  print("His name is " + name["name2"])

hemakesh(name1="pradeep",name2="rocky")
```

# DEFAULT PARAMETER VALUE

```python
def hemakesh(country ="Singapore"):
  print("I'm from " + country)

hemakesh("India")
hemakesh("Russia")
hemakesh()
hemakesh("Ausralia")
```

# PASSING LIST AS AN ARGUMENT

```python
def hemakesh(names):
  for x in names:
    print(x)

name=["Mahesh Babu","Prabhas","NTR"]
hemakesh(name)
```
