# Lists

## Creating a list 

```python
hemakesh=["red","orange","yellow","blue"]
print(hemakesh)
```

## Length of the list

```python
hemakesh=["red","orange","yellow","blue"]
print(len(hemakesh))
```

## Datatype of list

```python
hemakesh=["red","orange","yellow","blue"]
print(type(hemakesh))
```

## Range of indexes in list

```python
hemakesh=["red","orange","yellow","blue","black","white"]
print(hemakesh[2])
```

```python
hemakesh=["red","orange","yellow","blue","black","white"]
print(hemakesh[2:6])
```

```python
hemakesh=["red","orange","yellow","blue","black","white"]
print(hemakesh[-2])
```

```python
hemakesh=["red","orange","yellow","blue","black","white"]
print(hemakesh[3:])
```

```python
hemakesh=["red","orange","yellow","blue","black","white"]
print(hemakesh[:5])
```

```python
hemakesh=["red","orange","yellow","blue","black","white"]
print(hemakesh[-6:-3])
```

## To check whether the specific item is there in the list

```python
hemakesh=["red","orange","yellow","blue","black","white"]
if "black" in hemakesh:
  print("yes, it is there in the list")
else:
  print("no, it is not there in the list")
```

```python
hemakesh=["red","orange","yellow","blue","black","white"]
if "purple" in hemakesh:
  print("yes, it is there in the list")
else:
  print("no, it is not there in the list")
```

## Change of item in list

```python
hemakesh=["red","orange","yellow","blue","black","white"]
hemakesh[2]="RRR"
print(hemakesh)
```

```python
hemakesh=["red","orange","yellow","blue","black","white"]
hemakesh[2:5] = ["purple", "indigo","green"]
print(hemakesh)
```

## Insert items

```python
hemakesh=["red","orange","yellow","blue","black",]
hemakesh.insert(3,"white")
print(hemakesh)
```

## Append items

```python
hemakesh=["red","orange","yellow","blue","black","white"]
hemakesh.append("mahesh babu")
print(hemakesh)
```

## Extend lists

```python
hemakesh=["red","orange","yellow"]
mylist=["blue","black","white"]
hemakesh.extend(mylist)
print(hemakesh)
```

## Remove specified item

```python
hemakesh=["red","orange","yellow","blue","black","white"]
hemakesh.remove("blue")
print(hemakesh)
```

## Remove specified index

```python
hemakesh=["red","orange","yellow","blue","black","white"]
hemakesh.pop(3)
print(hemakesh)
```

## Clear the list

```python
hemakesh=["red","orange","yellow","blue","black","white"]
hemakesh.clear()
print(hemakesh)
```

## Loop through a list

```python
hemakesh=["red","orange","yellow","blue","black","white"]
for x in hemakesh:
  print(x)
```

## Loop through the index numbers

```python
hemakesh=["red","orange","yellow","blue","black","white"]
for x in range(len(hemakesh)):
  print(hemakesh[x])
```

## Using while loop

```python
hemakesh=["red","orange","yellow","blue","black","white"]
i=0
while i < len(hemakesh):
  print(hemakesh[i])
  i=i+1
```

## List comprehension

```python
hemakesh=["red","orange","yellow","blue","black","white"]
mylist=[]

for x in hemakesh:
  if "e" in x:
    mylist.append(x)

print(mylist)
```

```python
hemakesh=["red","orange","yellow","blue","black","white"]
mylist=[x for x in hemakesh if "e" in x]

print(mylist)
```

### Using condition

```python
hemakesh=["red","orange","yellow","blue","black","white"]
mylist = [x for x in hemakesh if x != "black"]
```

### Using Iterable

```python
hemakesh = [x for x in range(4)]

print(hemakesh)
```

```python
hemakesh = [x for x in range(10) if x < 4]

print(hemakesh)
```

### Using Expression

```python
hemakesh=["red","orange","yellow","blue","black","white"]
mylist = [x.upper() for x in hemakesh]

print(mylist)
```

```python
hemakesh=["red","orange","yellow","blue","black","white"]

mylist = [x if x != "orange" else "RRR" for x in hemakesh]

print(mylist)
```

## Sort lists alphanumberically

```python
hemakesh=["red","orange","yellow","blue","black","white"]
hemakesh.sort()
print(hemakesh)
```

```python
hemakesh = [95,25,35,19,64,76,15,84,36,10]
hemakesh.sort()
print(hemakesh)
```

## Sort descending

```python
hemakesh=["red","orange","yellow","blue","black","white"]
hemakesh.sort(reverse = True)
print(hemakesh)
```

```python
hemakesh = [95,25,35,19,64,76,15,84,36,10]
hemakesh.sort(reverse = True)
print(hemakesh)
```

## Case insensitive sort

```python
hemakesh=["red","orange","yellow","blue","black","white"]
hemakesh.sort()
print(hemakesh)
```

## Reverse order

```python
hemakesh=["red","orange","yellow","blue","black","white"]
hemakesh.reverse()
print(hemakesh)
```

## Copy a list

```python
hemakesh=["red","orange","yellow","blue","black","white"]
mylist = hemakesh.copy()
print(mylist)
```

```python
hemakesh=["red","orange","yellow","blue","black","white"]
mylist = list(hemakesh)
print(mylist)
```

## Join two lists

```python
hemakesh = ["red","orange","yellow","blue","black","white"]
mylist = [10, 11, 12, 13, 14, 15]

otherlist = hemakesh + mylist
print(otherlist)
```

```python
hemakesh = ["red","orange","yellow","blue","black","white"]
mylist = [10, 11, 12, 13, 14, 15]

for x in mylist:
  hemakesh.append(x)

print(hemakesh)
```

```python
hemakesh = ["red","orange","yellow","blue","black","white"]
mylist = [10, 11, 12, 13, 14, 15]

hemakesh.extend(mylist)
print(hemakesh)
```
